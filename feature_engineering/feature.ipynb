```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from xfeat import Pipeline, SelectCategorical, LambdaEncoder
from xfeat.num_encoder import SelectNumerical
```

```python
all_data = pd.read_csv('../data/all_data.csv')
```

```python
all_data.head(5)
```

```python
all_data.tail()
```

```python
#連続量を抜き出す
continuous_columns = all_data.select_dtypes(exclude=['object']).columns.tolist()
print(continuous_columns)

# 連続量のデータフレーム
all_data_continuous = pd.DataFrame(all_data[continuous_columns])
all_data_continuous.head()
```

```python
# 標準化したいカラムを指定
columns_to_standardize = ['市区町村コード', '建ぺい率（％）', '容積率（％）']

# 標準化
scaler = StandardScaler()
all_data[columns_to_standardize] = scaler.fit_transform(all_data[columns_to_standardize])
```

```python
# カテゴリ変数を抜き出す
categorical_columns = all_data.select_dtypes(include=['object']).columns.tolist()
print(categorical_columns)

# カテゴリ変数のデータフレーム
all_data_categorical = all_data[categorical_columns]
all_data_categorical.head()
```

```python
# ラベルエンコーディングを適用するカラムのリスト
columns_to_encode = ['種類', '都道府県名', '市区町村名', '地区名', '最寄駅：名称', '最寄駅：距離（分）', '間取り', '建築年', '建物の構造', '用途', '都市計画', '取引時点', '改装']

# LabelEncoderのインスタンスを作成
labelencoder = LabelEncoder()

# 指定された各カラムに対してラベルエンコーディングを適用
for column in columns_to_encode:
    all_data[column] = labelencoder.fit_transform(all_data[column])
```

```python
# '取引価格（総額）_log'がNaNでないデータを抽出
train_data = all_data[all_data['取引価格（総額）_log'].notna()]

# '取引価格（総額）_log'がNaNであるデータを抽出
test_data = all_data[all_data['取引価格（総額）_log'].isna()]

```
